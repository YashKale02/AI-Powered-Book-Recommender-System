```python
from statistics import correlation
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.heatmap(books.isnull().T, cbar=False, cmap='coolwarm')
plt.title("Missing Data Heatmap")
```

```python
import numpy as np

books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2024 - books["published_year"]
```

```python
columns_used = ["num_pages","average_rating","missing_description","age_of_book"]
correlation_matrix = books[columns_used].corr(method='spearman')

sns.set(style="whitegrid")
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Spearman Correlation Heatmap")
plt.show()
```

```python
book_missing = books.dropna(subset=["description", "num_pages", "average_rating", "published_year"])
```

```python
print(f"Original size: {books.shape[0]}")
print(f"Filtered size: {book_missing.shape[0]}")
```

```python
book_missing['categories'].value_counts()
```

```python
book_missing = book_missing.copy()
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
```

```python
book_missing.loc[book_missing["words_in_description"].between(15, 24), "description"]
```

```python
book_missing.loc[book_missing["words_in_description"].between(25, 34), "description"]
```

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
```

```python
book_missing_25_words
```

```python
book_missing_25_words = book_missing_25_words.copy()  #Create a deep copy to avoid SettingWithCopyWarning when modifying the DataFrame

# Combine title and subtitle into a single column; if subtitle is missing, just use the title
book_missing_25_words["title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(),
             book_missing_25_words["title"],
             book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
```

```python
book_missing_25_words
```

```python
# Combine ISBN and description into one string for each book (e.g., for tagging or embedding prep)
book_missing_25_words["tagged_description"] = (
    book_missing_25_words[["isbn13", "description"]]
    .astype(str)
    .agg(" ".join, axis=1)
)
```

```python
book_missing_25_words
```

```python
(
    book_missing_25_words
    .drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
    .to_csv("books_cleaned.csv", index = False)
)
```
