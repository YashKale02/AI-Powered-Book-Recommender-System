```python
# Import the pipeline function from Hugging Face Transformers
from transformers import pipeline
# categories you want to classify text into
fiction_categories = ["Fiction", "Nonfiction"]

# Create a zero-shot classification pipeline using the BART model
pipe = pipeline("zero-shot-classification",
                model="facebook/bart-large-mnli",   # Pretrained model name
                device=0)    # Device: 'mps' for Mac GPU, 0 for NVIDIA, '-1' for CPU
```

```python
#extracts the description of the first book in the dataset that has its simple_categories marked as "Fiction".
#books.loc to select particular data, .reset_index to reset index of data as it can be non-continuous, 0 to get the first one data which pass condition
sequence = books.loc[books["simple_categories"] == "Fiction", "description"].reset_index(drop=True)[0]
```

```python
# Classify the sequence as either "Fiction" or "Nonfiction" using zero-shot classification
pipe(sequence, fiction_categories)
```

```python
import numpy as np

#This finds the category label with the highest confidence score from the model's prediction.
#np.argmax() finds the index of the highest score in the list:fiction/nonfiction(0,1)
max_index = np.argmax(pipe(sequence, fiction_categories)["scores"])
#now take label array and return the index with max probability using the max index
max_label = pipe(sequence, fiction_categories)["labels"][max_index]
max_label
```

```python
#jus a function for above task and giving categories column as argument
def generate_predictions(sequence, categories):
    predictions = pipe(sequence, categories)
    max_index = np.argmax(predictions["scores"])
    max_label = predictions["labels"][max_index]
    return max_label
```

```python
#This loop evaluates the model's prediction on the first 300 "Fiction" book descriptions, storing the actual and predicted categories for later comparison
from tqdm import tqdm #tells how much data left to process real time

actual_cats = []
predicted_cats = []

for i in tqdm(range(300, 600)):
    sequence = books.loc[books["simple_categories"] == "Fiction", "description"].reset_index(drop=True)[i]
    predicted_cats += [generate_predictions(sequence, fiction_categories)]
    actual_cats += ["Fiction"]
#means in reality all the categories that model is predicting on is fiction but we're checking how correctly it predicts
```

```python
for i in tqdm(range(300, 600)):
    sequence = books.loc[books["simple_categories"] == "Nonfiction", "description"].reset_index(drop=True)[i]
    predicted_cats += [generate_predictions(sequence, fiction_categories)]
    actual_cats += ["Nonfiction"]
```

```python
predictions_df = pd.DataFrame({"actual_categories": actual_cats, "predicted_categories": predicted_cats})
```

```python
predictions_df
```

```python
predictions_df["correct_prediction"] = (
    np.where(predictions_df["actual_categories"] == predictions_df["predicted_categories"], 1, 0)
)
#created new column which has 0,1 value based on the condition
```

```python
predictions_df["correct_prediction"].sum() / len(predictions_df)
#tells accuracy by sum of all 1(right pred.)/ total predictions
```

```python
isbns = []             # To store ISBNs of books with missing categories
predicted_cats = []    # To store predicted categories for those books

# Filter books where 'simple_categories' is missing (NaN), keep only ISBN and description
missing_cats = books.loc[books["simple_categories"].isna(), ["isbn13", "description"]].reset_index(drop=True)
```

```python
for i in tqdm(range(0, len(missing_cats))):  # Loop through books with missing categories
    sequence = missing_cats["description"][i]             # Get book description
    predicted_cats += [generate_predictions(sequence, fiction_categories)]  # Predict category
    isbns += [missing_cats["isbn13"][i]]                  # Store corresponding ISBN
```

```python
# Create a new DataFrame with ISBNs and their predicted categories
missing_predicted_df = pd.DataFrame({
    "isbn13": isbns,
    "predicted_categories": predicted_cats
})
```

```python
missing_predicted_df
```

```python
# Merge predictions back into the main books DataFrame on isbn13
books = pd.merge(books, missing_predicted_df, on="isbn13", how="left")

# If 'simple_categories' is missing (NaN), fill it with predicted category
books["simple_categories"] = np.where(books["simple_categories"].isna(), books["predicted_categories"], books["simple_categories"])

# Drop the now-unneeded 'predicted_categories' column
books = books.drop(columns=["predicted_categories"])
```
